##  PIE

```python
x = np.array([5,7,8,7,2])
plt.pie(x)
plt.show()
```

```python
x = np.array([5,7,8,7,2])

labels = ["apple" , "banana" , "cherries" , "grapes" , "strawbwerry"]
myexplode = [0,0.2,0,0,0]
plt.pie(x , labels=labels, explode=myexplode,startangle=90,shadow=True)
plt.show()
```

```python
x = np.array([5,7,8,7,2])

labels = ["apple" , "banana" , "cherries" , "grapes" , "strawbwerry"]
myexplode = [0,0.2,0,0,0]
c =['red' , 'green' ,'blue' , 'yellow' , 'black','pink']
plt.pie(x , labels=labels, explode=myexplode,startangle=90,shadow=True)
plt.show()
```

```python
x = np.array([5,7,8,7,2])

labels = ["apple" , "banana" , "cherries" , "grapes" , "strawbwerry"]
myexplode = [0,0.2,0,0,0]

c =['red' , 'green' ,'blue' , 'yellow' , 'black','pink']
plt.pie(x , labels=labels, explode=myexplode,startangle=90,shadow=True, colors=c,autopct='%1.1f%%')
plt.legend(title = "my pie chart")
plt.show()
```
